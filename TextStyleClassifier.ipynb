```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_excel("1.xlsx")
X_data = df1['text'].values
X_data = X_data.tolist()
target = df1['class'].values

df2 = pd.read_excel("2.xlsx")
X_new = df2['text'].values
print(target)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(target)
target = encoder.transform(target)
```

```python
print(X_data)
print(target)
print(df2)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import tqdm


def preprocess_text(texts):
    stop_words = set(stopwords.words('russian'))
    print(stop_words)
    regex = re.compile('[^а-я А-Я]')
    preprocess_texts = []
    for i in  tqdm.tqdm(range(len(texts))):
        text = texts[i].lower()
        text = regex.sub(' ', text)
        word_tokens = word_tokenize(text) 
        filtered_sentence = [w for w in word_tokens if not w in stop_words] 
        preprocess_texts.append( ' '.join(filtered_sentence))
    
    return preprocess_texts
```

```python
X_data_stop = preprocess_text(X_data)
print(X_data_stop)
print(len(X_data_stop))
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
def stemming_texts(texts):
  st = SnowballStemmer('russian')
  stem_text = []
  for text in tqdm.tqdm(texts):
    word_tokens = word_tokenize(text)
    stem_text.append(' '.join([st.stem(word) for word in word_tokens]))
  return stem_text
```

```python
X_data_stem = stemming_texts(X_data_stop)
print(X_data_stem)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,3))
X_data = vectorizer.fit_transform(X_data)
 
X_train, X_test,y_train,y_test = train_test_split(
                    X_data,          # обучающие данные
                    target,          # ответы
                    train_size=0.75, # 75 % на обучение, 25% - на тест точности 
                    random_state=10)
```

```python
print(X_train)
print(X_test)
print(X_train.shape)
```

```python
clf = KNeighborsClassifier(n_neighbors = 3) # K-ближайших соседей
clf.fit(X_train, y_train)
# прогноз для тестового набора
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
#print(y_pred)
print(accuracy_score(y_train_pred, y_train))
print(accuracy_score(y_test_pred, y_test))
```
