```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_excel("1.xlsx")
X_data = df1['text'].values
X_data = X_data.tolist()
target = df1['class'].values

df2 = pd.read_excel("2.xlsx")
X_new = df2['text'].values
print(target)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(target)
target = encoder.transform(target)
```

```python
print(X_data)
print(target)
print(df2)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,3))
X_data = vectorizer.fit_transform(X_data)
 
X_train, X_test,y_train,y_test = train_test_split(
                    X_data,          # обучающие данные
                    target,          # ответы
                    train_size=0.75, # 75 % на обучение, 25% - на тест точности 
                    random_state=10)
```

```python
print(X_train)
print(X_test)
```

```python
clf = KNeighborsClassifier(n_neighbors = 3) # K-ближайших соседей
clf.fit(X_train, y_train)
# прогноз для тестового набора
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
#print(y_pred)
print(accuracy_score(y_train_pred, y_train))
print(accuracy_score(y_test_pred, y_test))
```

```python
X_new_vec = vectorizer.transform(X_new) #
# прогноз для нового документа
X_new_pred = clf.predict(X_new_vec)
print(X_new_pred)
```

```python
from sklearn.linear_model import LogisticRegression

regr = LogisticRegression()
regr.fit(X_train, y_train)
y_train_pred_regr = regr.predict(X_train)
y_test_pred_regr = regr.predict(X_test)
```

```python
print(accuracy_score(y_train_pred_regr, y_train))
print(accuracy_score(y_test_pred_regr, y_test))
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)
y_test_pred_svm = svm.predict(X_test)
```

```python
print(accuracy_score(y_test_pred_svm, y_test))
```
